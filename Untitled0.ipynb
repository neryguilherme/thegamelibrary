```python


import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

banco=pd.read_parquet('/content/games_cleaned.parquet')
banco['id'] = range(1, len(banco) + 1)

# Preencher valores faltantes
for col in banco.columns:
    if banco[col].dtype == 'object':
        banco[col] = banco[col].fillna('missing')
    elif pd.api.types.is_numeric_dtype(banco[col]):
        banco[col] = banco[col].fillna(banco[col].median())

# Separar a coluna 'name' das demais
name_column = banco['Name']
x_column = banco.drop('Name', axis=1)

# Agora vocÃª tem duas variÃ¡veis:
# - name_column: ContÃ©m apenas a coluna 'name'
# - other_columns: ContÃ©m todas as outras colunas do DataFrame

# Exemplo de uso:
print("Coluna 'name':\n", name_column.head())
print("\nOutras colunas:\n", x_column.head())
```

```python


from sklearn.preprocessing import LabelEncoder

# Criar um LabelEncoder
le = LabelEncoder()

# Aplicar o LabelEncoder a cada coluna categÃ³rica
for col in x_column.columns:
    if x_column[col].dtype == 'object':
        x_column[col] = le.fit_transform(x_column[col])

# Agora x_column contÃ©m apenas valores numÃ©ricos
print(x_column.head())
```

```python


from sklearn.preprocessing import StandardScaler

# Crie um StandardScaler
scaler = StandardScaler()

# Aplique o scaler Ã s colunas numÃ©ricas de x_column
numeric_cols = x_column.select_dtypes(include=np.number).columns
x_column[numeric_cols] = scaler.fit_transform(x_column[numeric_cols])

# Agora x_column tem as colunas numÃ©ricas padronizadas
print(x_column.head())
```

```python


import pandas as pd
import numpy as np

# ... (your existing code)

# Calculate the variance of each column
variances = x_column.var()

# Set a threshold for low variance (you can adjust this value)
threshold = 0.5

# Identify columns with low variance
low_variance_cols = variances[variances < threshold].index

# Print the names of the columns with low variance
print("Removed Columns (Low Variance):", low_variance_cols.tolist())

# Remove columns with low variance from the DataFrame
x_column = x_column.drop(columns=low_variance_cols)
x_column.head()
```

```python


from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

# Supondo que 'x_column_filtrado' e 'name_column' estejam definidos como no cÃ³digo anterior

# Divida os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(x_column, name_column, test_size=0.3, random_state=42)

# Crie o classificador MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(20,20), max_iter=500, alpha=0.0001,
                     solver='adam', verbose=10,  random_state=21,tol=0.0001)

# Treine o classificador
mlp.fit(X_train, y_train)

# FaÃ§a previsÃµes no conjunto de teste
y_pred = mlp.predict(X_test)

# Avalie o modelo (adicione mÃ©tricas conforme necessÃ¡rio)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f"AcurÃ¡cia: {accuracy}")
```

```python


from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Selecionar as colunas relevantes
features = ['Positive', 'Genres', 'Price']
target = 'Name'  # Assumindo que 'Name' representa o jogo a ser recomendado

# Criar o dataset para o modelo
X = x_column[features]
y = name_column

# Dividir os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=42)

# Inicializar o classificador MLP
mlp = MLPClassifier(max_iter=100, verbose=True, tol=0.0000100,solver = 'adam', activation = 'relu',hidden_layer_sizes = (5,5))

# Treinar o modelo
mlp.fit(X_train, y_train)

# Fazer previsÃµes
y_pred = mlp.predict(X_test)

# Avaliar a precisÃ£o do modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"AcurÃ¡cia do modelo: {accuracy}")


# FunÃ§Ã£o para recomendar jogos
def recomendar_jogos(positive, genre, price, n=5):
    # Criar um DataFrame com os dados de entrada
    novo_jogo = pd.DataFrame({'Positive': [positive], 'Genres': [genre], 'Price': [price]})

    # Aplicar as mesmas transformaÃ§Ãµes que foram aplicadas aos dados de treino
    for col in novo_jogo.columns:
      if novo_jogo[col].dtype == 'object':
        novo_jogo[col] = le.transform(novo_jogo[col])
    numeric_cols_novo = novo_jogo.select_dtypes(include=np.number).columns
    novo_jogo[numeric_cols_novo] = scaler.transform(novo_jogo[numeric_cols_novo])


    # Obter as previsÃµes do modelo
    previsoes = mlp.predict_proba(novo_jogo)

    # Obter os Ã­ndices dos jogos mais provÃ¡veis
    top_indices = np.argsort(previsoes[0])[::-1][:n]

    # Retornar os nomes dos jogos mais provÃ¡veis
    nomes_jogos = [le.inverse_transform([int(i)])[0] for i in top_indices ]
    return nomes_jogos


# Exemplo de uso da funÃ§Ã£o de recomendaÃ§Ã£o (precisa ajustar os valores para as features)
recomendacoes = recomendar_jogos(positive=80, genre='Action', price=19.99)
print("\nRecomendaÃ§Ãµes:")
recomendacoes
```
